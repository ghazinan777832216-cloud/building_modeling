# Python GIS Procedural Pipeline

This notebook implements an end-to-end pipeline:
1. Draw a polygon on an interactive map
2. Extract GeoJSON & transform coordinates to UTM (CCW ordering)
3. Generate a procedural building with PyPRT (CityEngine)
4. Visualize the 3D model with Three.js Viewer + Interactive Map

## الخلية 1: إعداد الخريطة ورسم المضلع
## Cell 1: Interactive Map & Polygon Drawing

````python
import leafmap

m = leafmap.Map(center=[15, 45], zoom=16)

# The draw tool is available by default in leafmap.
# Draw a polygon on the map below, then proceed to Cell 2.
m
````

---
## الخلية 2: استخراج GeoJSON وتحويل الإحداثيات إلى UTM وترتيب CCW
## Cell 2: Extract GeoJSON → Transform to UTM → Ensure CCW Winding

````python
from shapely.geometry import shape, Polygon
from pyproj import Transformer

if len(m.draw_features) == 0:
    raise RuntimeError("لم يتم رسم أي مضلع بعد — No polygon drawn yet. Run Cell 1 and draw a polygon first.")

# Take the most recently drawn feature
geojson_feature = m.draw_features[-1]
polygon = shape(geojson_feature['geometry'])
coords = list(polygon.exterior.coords)

print("إحداثيات Lat/Lon الأصلية (Original Lat/Lon coordinates):")
print(coords)

# Transform from WGS84 (EPSG:4326) to UTM Zone 38N (EPSG:32638)
transformer = Transformer.from_crs("EPSG:4326", "EPSG:32638", always_xy=True)
utm_coords = [transformer.transform(x, y) for x, y in coords]
utm_polygon = Polygon(utm_coords)

# Ensure counter-clockwise (CCW) winding order
if not utm_polygon.exterior.is_ccw:
    utm_coords = list(utm_polygon.exterior.coords)[::-1]
    utm_polygon = Polygon(utm_coords)

print("\nتم تحويل الإحداثيات إلى UTM وترتيبها CCW بنجاح")
print("(Coordinates transformed to UTM and ordered CCW successfully)")
print(f"\nUTM Coords ({len(utm_coords)} vertices):")
for i, c in enumerate(utm_coords):
    print(f"  [{i}] x={c[0]:.2f}, y={c[1]:.2f}")

from IPython.display import display
display(utm_polygon)  # Show the 2D polygon shape

# Create an interactive map for the extracted feature
m2 = leafmap.Map(center=[15, 45], zoom=16)
fc = {"type": "FeatureCollection", "features": [geojson_feature]}
m2.add_geojson(fc, layer_name="Extracted Polygon")
m2
````

---
## الخلية 3: إرسال الشكل إلى PyPRT وتوليد المبنى Procedural
## Cell 3: Send Shape to PyPRT & Generate Procedural Building

````python
import pyprt, os, base64
import numpy as np

pyprt.initialize_prt()

# إخراج الإحداثيات للمبنى مع التوسيط (Centering) لضمان ظهوره في الفيوور
coords_2d = list(utm_polygon.exterior.coords)[:-1]

# حساب المركز (Centroid) لجعل الموديل حول نقطة الصفر (0,0,0)
center_x = sum(c[0] for c in coords_2d) / len(coords_2d)
center_y = sum(c[1] for c in coords_2d) / len(coords_2d)

flat_coords = []
for x, y in coords_2d:
    # طرح المركز لضمان التوسيط التام
    flat_coords.extend([x - center_x, 0.0, y - center_y])

initial_shape = pyprt.InitialShape(flat_coords)
attributes = {"Nbr_of_Floors": 12, "Usage": "Residential"}

RPK_PATH = r"C:\RPK\RuleFootprint.rpk"
model_generator = pyprt.ModelGenerator([initial_shape])

encoder = "com.esri.prt.codecs.GLTFEncoder"
out_dir = os.path.join(os.getcwd(), "output_models")
os.makedirs(out_dir, exist_ok=True)

encoder_options = {
    "outputPath": out_dir,
    "baseName": "procedural_building"
}

# توليد المبنى
model_generator.generate_model([attributes], RPK_PATH, encoder, encoder_options)

# قراءة الملف وتحويله لـ Base64
glb_path = os.path.join(out_dir, "procedural_building_0.glb")
with open(glb_path, "rb") as f:
    glb_data = f.read()

model_b64 = base64.b64encode(glb_data).decode('utf-8')

print(f"✅ Model Centered and Ready!")
print(f"Shifted by Central Offset: X={center_x:.2f}, Y={center_y:.2f}")
````

````python
import os, webbrowser
from IPython.display import display, HTML

html_content = '''<!DOCTYPE html>
<html>
<head>
    <title>CityEngine 3D Viewer + Map + Drawing</title>
    <meta charset="utf-8">
    <link rel="stylesheet" href="https://unpkg.com/leaflet@1.9.4/dist/leaflet.css" />
    <link rel="stylesheet" href="https://unpkg.com/leaflet-draw@1.0.4/dist/leaflet.draw.css" />
    <style>
        * { margin: 0; padding: 0; box-sizing: border-box; }
        body { overflow: hidden; background: #1a1a2e; font-family: "Segoe UI", sans-serif; }
        #container { display: flex; width: 100vw; height: 100vh; }
        #viewer-panel { flex: 1; position: relative; }
        #map-panel { width: 400px; position: relative; border-left: 3px solid #0076d1; }
        #map { width: 100%; height: 100%; }
        #toggle-map {
            position: absolute; top: 50%; right: 0; transform: translateY(-50%);
            z-index: 1000; background: #0076d1; color: white; border: none;
            padding: 20px 8px; cursor: pointer; font-size: 16px;
            border-radius: 8px 0 0 8px; box-shadow: -2px 0 10px rgba(0,0,0,0.3);
        }
        #toggle-map:hover { background: #005ba1; }
        #toolbar {
            position: absolute; top: 15px; left: 15px; z-index: 10;
            display: flex; flex-direction: column; gap: 6px;
        }
        .btn {
            background: rgba(0,118,209,0.9); color: white; border: none;
            padding: 8px 14px; border-radius: 5px; cursor: pointer;
            font-weight: 600; font-size: 12px;
            box-shadow: 0 2px 8px rgba(0,0,0,0.3);
        }
        .btn:hover { background: rgba(0,90,161,0.95); }
        .btn.gray { background: rgba(80,80,80,0.85); }
        #status {
            background: rgba(0,0,0,0.85); color: #0f0;
            padding: 8px 12px; border-radius: 5px;
            font-family: Consolas, monospace; font-size: 11px;
            min-width: 260px; border-left: 3px solid #0f0;
        }
        #info-bar {
            position: absolute; bottom: 10px; left: 10px; z-index: 10;
            background: rgba(0,0,0,0.7); color: #aaa; padding: 6px 12px;
            border-radius: 5px; font-size: 10px; font-family: Consolas, monospace;
        }
        #map-header {
            position: absolute; top: 0; left: 0; right: 0; z-index: 1000;
            background: linear-gradient(180deg, rgba(0,118,209,0.95), rgba(0,118,209,0.8));
            color: white; padding: 10px 15px; font-weight: 600; font-size: 14px;
            text-align: center;
        }
        #draw-tools-3d {
            position: absolute; top: 15px; right: 15px; z-index: 10;
            display: flex; flex-direction: column; gap: 6px;
        }
        .draw-btn {
            background: rgba(156,39,176,0.9); color: white; border: none;
            padding: 8px 14px; border-radius: 5px; cursor: pointer;
            font-weight: 600; font-size: 12px;
            box-shadow: 0 2px 8px rgba(0,0,0,0.3);
        }
        .draw-btn:hover { background: rgba(123,31,162,0.95); }
        .draw-btn.active { background: rgba(255,152,0,0.9); box-shadow: 0 0 10px rgba(255,152,0,0.5); }
    </style>
</head>
<body>
    <div id="container">
        <div id="viewer-panel">
            <div id="toolbar">
                <div id="status">Loading...</div>
                <button class="btn" onclick="resetCamera()">🎯 Focus Building</button>
                <button class="btn gray" onclick="toggleWireframe()">📐 Wireframe</button>
                <button class="btn gray" onclick="toggleAutoRotate()">🔄 Auto Rotate</button>
                <button class="btn gray" onclick="toggleGrid()">🔲 Grid</button>
            </div>
            <div id="draw-tools-3d">
                <button class="draw-btn" id="btn-draw-poly" onclick="startDrawPolygon()">🔷 Draw Polygon</button>
                <button class="draw-btn" id="btn-draw-rect" onclick="startDrawRect()">⬜ Draw Rectangle</button>
                <button class="draw-btn" id="btn-draw-line" onclick="startDrawLine()">📏 Draw Line</button>
                <button class="draw-btn" id="btn-measure" onclick="startMeasure()">📐 Measure</button>
                <button class="draw-btn" onclick="clearDrawings()">🗑 Clear All</button>
                <button class="draw-btn" onclick="finishDrawing()" id="btn-finish" style="display:none">✅ Finish</button>
            </div>
            <div id="info-bar">Left=Rotate | Scroll=Zoom | Right=Pan</div>
            <button id="toggle-map" onclick="toggleMapPanel()">🗺</button>
        </div>
        <div id="map-panel">
            <div id="map-header">🗺 Interactive Map</div>
            <div id="map"></div>
        </div>
    </div>

    <script src="https://unpkg.com/leaflet@1.9.4/dist/leaflet.js"></script>
    <script src="https://unpkg.com/leaflet-draw@1.0.4/dist/leaflet.draw.js"></script>
    <script>
    var mapObj = L.map("map", {zoomControl:true}).setView([15, 45], 16);
    var sat = L.tileLayer("https://server.arcgisonline.com/ArcGIS/rest/services/World_Imagery/MapServer/tile/{z}/{y}/{x}", {attribution:"Esri"});
    var osm = L.tileLayer("https://{s}.tile.openstreetmap.org/{z}/{x}/{y}.png", {attribution:"OSM"});
    sat.addTo(mapObj);
    L.control.layers({"Satellite":sat, "OpenStreetMap":osm}).addTo(mapObj);
    var drawnItems = new L.FeatureGroup(); mapObj.addLayer(drawnItems);
    mapObj.addControl(new L.Control.Draw({
        draw: { polygon:{shapeOptions:{color:"#ff6600",weight:3}}, rectangle:{shapeOptions:{color:"#ff6600",weight:3}}, polyline:{shapeOptions:{color:"#0088ff",weight:3}}, circle:false, circlemarker:false, marker:true },
        edit: { featureGroup: drawnItems }
    }));
    mapObj.on(L.Draw.Event.CREATED, function(e){ drawnItems.addLayer(e.layer); });
    var mapVisible = true;
    window.toggleMapPanel = function() {
        var p = document.getElementById("map-panel");
        mapVisible = !mapVisible;
        p.style.display = mapVisible ? "block" : "none";
        document.getElementById("toggle-map").textContent = mapVisible ? "🗺" : "📍";
        if(mapVisible) setTimeout(function(){mapObj.invalidateSize();},100);
        window.dispatchEvent(new Event("resize"));
    };
    </script>

    <script type="importmap">
    { "imports": { "three": "https://cdn.jsdelivr.net/npm/three@0.160.0/build/three.module.js", "three/addons/": "https://cdn.jsdelivr.net/npm/three@0.160.0/examples/jsm/" } }
    </script>

    <script type="module">
    import * as THREE from "three";
    import { OrbitControls } from "three/addons/controls/OrbitControls.js";
    import { GLTFLoader } from "three/addons/loaders/GLTFLoader.js";

    var log = function(m){ document.getElementById("status").innerHTML = m; };
    var wireframeMode = false, loadedModel = null, gridVisible = true;
    var viewerPanel = document.getElementById("viewer-panel");

    var scene = new THREE.Scene();
    scene.background = new THREE.Color(0x87CEEB);
    scene.fog = new THREE.Fog(0x87CEEB, 800, 3000);

    var camera = new THREE.PerspectiveCamera(45, viewerPanel.clientWidth/viewerPanel.clientHeight, 0.1, 5000);
    camera.position.set(150, 200, 250);

    var renderer = new THREE.WebGLRenderer({antialias:true});
    renderer.setSize(viewerPanel.clientWidth, viewerPanel.clientHeight);
    renderer.setPixelRatio(window.devicePixelRatio);
    renderer.shadowMap.enabled = true;
    renderer.shadowMap.type = THREE.PCFSoftShadowMap;
    renderer.toneMapping = THREE.ACESFilmicToneMapping;
    renderer.toneMappingExposure = 1.2;
    viewerPanel.appendChild(renderer.domElement);

    var controls = new OrbitControls(camera, renderer.domElement);
    controls.enableDamping = true; controls.dampingFactor = 0.05;
    controls.target.set(0, -76, 0);

    scene.add(new THREE.AmbientLight(0xffffff, 0.6));
    var dL = new THREE.DirectionalLight(0xffffff, 1.2);
    dL.position.set(100,200,100); dL.castShadow = true;
    dL.shadow.mapSize.set(2048,2048);
    dL.shadow.camera.left=-200; dL.shadow.camera.right=200;
    dL.shadow.camera.top=200; dL.shadow.camera.bottom=-200;
    scene.add(dL);
    scene.add(new THREE.HemisphereLight(0x87CEEB, 0x362D1B, 0.5));

    var gGeo = new THREE.PlaneGeometry(2000,2000);
    var gMat = new THREE.MeshStandardMaterial({color:0x4a7c59, roughness:0.9});
    var ground = new THREE.Mesh(gGeo, gMat);
    ground.rotation.x = -Math.PI/2; ground.receiveShadow = true;
    scene.add(ground);

    var gridHelper = new THREE.GridHelper(500,50,0x888888,0x444444);
    gridHelper.position.y = 0.1; scene.add(gridHelper);
    scene.add(new THREE.AxesHelper(50));

    // 3D Drawing
    var drawMode = null, drawPoints = [], drawnShapes3D = [];
    var drawGroup = new THREE.Group(); scene.add(drawGroup);
    var raycaster = new THREE.Raycaster(), mouse = new THREE.Vector2();

    function getGround(e){
        var r = renderer.domElement.getBoundingClientRect();
        mouse.x = ((e.clientX-r.left)/r.width)*2-1;
        mouse.y = -((e.clientY-r.top)/r.height)*2+1;
        raycaster.setFromCamera(mouse, camera);
        var h = raycaster.intersectObject(ground);
        return h.length > 0 ? h[0].point : null;
    }
    function addDot(pt){
        var d = new THREE.Mesh(new THREE.SphereGeometry(1.5), new THREE.MeshBasicMaterial({color:0xff6600}));
        d.position.copy(pt); d.position.y += 0.5; drawGroup.add(d);
        drawPoints.push(pt.clone());
        if(drawPoints.length > 1){
            var pts = drawPoints.slice(-2).map(function(p){return new THREE.Vector3(p.x,p.y+0.5,p.z);});
            drawGroup.add(new THREE.Line(new THREE.BufferGeometry().setFromPoints(pts), new THREE.LineBasicMaterial({color:0xff6600})));
        }
        log("📍 Point " + drawPoints.length + " placed");
    }
    function makePolygon(pts){
        if(pts.length < 3) return;
        var s = new THREE.Shape();
        s.moveTo(pts[0].x, pts[0].z);
        for(var i=1;i<pts.length;i++) s.lineTo(pts[i].x, pts[i].z);
        s.lineTo(pts[0].x, pts[0].z);
        var m = new THREE.Mesh(new THREE.ShapeGeometry(s), new THREE.MeshBasicMaterial({color:0xff6600,transparent:true,opacity:0.3,side:THREE.DoubleSide}));
        m.rotation.x = -Math.PI/2; m.position.y = ground.position.y+1;
        drawGroup.add(m); drawnShapes3D.push(m);
        var cl = pts.concat([pts[0]]).map(function(p){return new THREE.Vector3(p.x,ground.position.y+0.5,p.z);});
        drawGroup.add(new THREE.Line(new THREE.BufferGeometry().setFromPoints(cl), new THREE.LineBasicMaterial({color:0xff6600})));
    }

    renderer.domElement.addEventListener("click", function(e){
        if(!drawMode) return;
        var pt = getGround(e); if(!pt) return;
        if(drawMode==="measure" && drawPoints.length>=1){
            addDot(pt); var dist=drawPoints[0].distanceTo(drawPoints[1]);
            log("📐 Distance: <b>"+dist.toFixed(2)+" m</b>");
            drawMode=null; document.getElementById("btn-measure").classList.remove("active"); drawPoints=[];
            return;
        }
        if(drawMode==="rect" && drawPoints.length===0){ addDot(pt); log("⬜ Click opposite corner"); }
        else if(drawMode==="rect" && drawPoints.length===1){
            var p1=drawPoints[0], p2=pt;
            var rp=[new THREE.Vector3(p1.x,p1.y,p1.z),new THREE.Vector3(p2.x,p1.y,p1.z),new THREE.Vector3(p2.x,p2.y,p2.z),new THREE.Vector3(p1.x,p2.y,p2.z)];
            drawPoints=rp; makePolygon(rp); log("⬜ Rectangle drawn!");
            drawMode=null; document.getElementById("btn-draw-rect").classList.remove("active"); drawPoints=[];
        } else { addDot(pt); }
        document.getElementById("btn-finish").style.display = drawMode ? "block" : "none";
    });

    window.startDrawPolygon = function(){ drawMode="polygon"; drawPoints=[];
        document.querySelectorAll(".draw-btn").forEach(function(b){b.classList.remove("active");});
        document.getElementById("btn-draw-poly").classList.add("active");
        document.getElementById("btn-finish").style.display = "block";
        log("🔷 Click points on ground, then Finish"); };
    window.startDrawRect = function(){ drawMode="rect"; drawPoints=[];
        document.querySelectorAll(".draw-btn").forEach(function(b){b.classList.remove("active");});
        document.getElementById("btn-draw-rect").classList.add("active");
        log("⬜ Click first corner"); };
    window.startDrawLine = function(){ drawMode="line"; drawPoints=[];
        document.querySelectorAll(".draw-btn").forEach(function(b){b.classList.remove("active");});
        document.getElementById("btn-draw-line").classList.add("active");
        document.getElementById("btn-finish").style.display = "block";
        log("📏 Click points, then Finish"); };
    window.startMeasure = function(){ drawMode="measure"; drawPoints=[];
        document.querySelectorAll(".draw-btn").forEach(function(b){b.classList.remove("active");});
        document.getElementById("btn-measure").classList.add("active");
        log("📐 Click two points to measure"); };
    window.finishDrawing = function(){
        if(drawMode==="polygon" && drawPoints.length>=3){ makePolygon(drawPoints); log("🔷 Polygon done! " + drawPoints.length + " vertices"); }
        else if(drawMode==="line"){ log("📏 Line done! " + drawPoints.length + " points"); }
        drawMode=null; drawPoints=[];
        document.querySelectorAll(".draw-btn").forEach(function(b){b.classList.remove("active");});
        document.getElementById("btn-finish").style.display = "none"; };
    window.clearDrawings = function(){
        while(drawGroup.children.length>0) drawGroup.remove(drawGroup.children[0]);
        drawnShapes3D.length=0; drawPoints=[]; drawMode=null;
        document.querySelectorAll(".draw-btn").forEach(function(b){b.classList.remove("active");});
        document.getElementById("btn-finish").style.display = "none";
        log("🗑 Cleared"); };

    // Load GLB
    log("📦 Loading model...");
    var b64 = "%%MODEL_B64%%";
    var bin = atob(b64), bytes = new Uint8Array(bin.length);
    for(var i=0;i<bin.length;i++) bytes[i]=bin.charCodeAt(i);
    var blob = new Blob([bytes.buffer], {type:"model/gltf-binary"});
    var url = URL.createObjectURL(blob);

    new GLTFLoader().load(url, function(gltf){
        loadedModel = gltf.scene;
        loadedModel.traverse(function(c){
            if(c.isMesh){ c.castShadow=true; c.receiveShadow=true; if(c.material) c.material.side=THREE.DoubleSide; }
        });
        scene.add(loadedModel);
        var box = new THREE.Box3().setFromObject(loadedModel);
        var center = box.getCenter(new THREE.Vector3());
        var size = box.getSize(new THREE.Vector3());
        var maxDim = Math.max(size.x, size.y, size.z);
        camera.position.set(center.x+maxDim, center.y+maxDim*0.7, center.z+maxDim);
        controls.target.copy(center); controls.update();
        ground.position.y = box.min.y;
        gridHelper.position.y = box.min.y + 0.1;
        log("✅ <span style=color:#4caf50>Building Ready!</span> " + size.x.toFixed(0)+"x"+size.y.toFixed(0)+"x"+size.z.toFixed(0)+"m");
        URL.revokeObjectURL(url);
    }, null, function(err){ log("❌ " + err.message); });

    window.resetCamera = function(){
        if(!loadedModel) return;
        var box=new THREE.Box3().setFromObject(loadedModel), c=box.getCenter(new THREE.Vector3()), s=box.getSize(new THREE.Vector3()), m=Math.max(s.x,s.y,s.z);
        camera.position.set(c.x+m, c.y+m*0.7, c.z+m); controls.target.copy(c); controls.update();
        log("🎯 Focused"); };
    window.toggleWireframe = function(){
        wireframeMode=!wireframeMode;
        if(loadedModel) loadedModel.traverse(function(c){if(c.isMesh&&c.material) c.material.wireframe=wireframeMode;});
        log(wireframeMode?"📐 Wireframe ON":"📐 Wireframe OFF"); };
    window.toggleAutoRotate = function(){
        controls.autoRotate=!controls.autoRotate; controls.autoRotateSpeed=2.0;
        log(controls.autoRotate?"🔄 Rotating":"🔄 Stopped"); };
    window.toggleGrid = function(){
        gridVisible=!gridVisible; gridHelper.visible=gridVisible;
        log(gridVisible?"🔲 Grid ON":"🔲 Grid OFF"); };

    window.addEventListener("resize", function(){
        var w=viewerPanel.clientWidth, h=viewerPanel.clientHeight;
        camera.aspect=w/h; camera.updateProjectionMatrix(); renderer.setSize(w,h);
    });
    (function animate(){ requestAnimationFrame(animate); controls.update(); renderer.render(scene,camera); })();
    </script>
</body>
</html>'''

final_html = html_content.replace('%%MODEL_B64%%', model_b64)
html_filename = 'viewer_full.html'
with open(html_filename, 'w', encoding='utf-8') as f:
    f.write(final_html)

full_path = os.path.abspath(html_filename)
webbrowser.open('file:///' + full_path.replace(chr(92), '/'))

display(HTML("""
<div style='border: 2px solid #28a745; background: linear-gradient(135deg, #f0fff4, #e8f5e9); padding: 25px; border-radius: 12px; font-family: system-ui; text-align: center;'>
    <h2 style='color: #28a745; margin-top: 0;'>🎉 الفيوور الكامل جاهز!</h2>
    <p style='color: #333;'>يحتوي على أدوات رسم ثلاثية الأبعاد + خريطة تفاعلية في لوحة منفصلة</p>
    <div style='display: flex; gap: 10px; justify-content: center; flex-wrap: wrap; margin: 15px 0;'>
        <span style='background: #e3f2fd; padding: 6px 12px; border-radius: 20px; font-size: 12px;'>🔷 رسم مضلع</span>
        <span style='background: #e3f2fd; padding: 6px 12px; border-radius: 20px; font-size: 12px;'>⬜ رسم مستطيل</span>
        <span style='background: #e3f2fd; padding: 6px 12px; border-radius: 20px; font-size: 12px;'>📏 رسم خط</span>
        <span style='background: #e3f2fd; padding: 6px 12px; border-radius: 20px; font-size: 12px;'>📐 قياس المسافة</span>
        <span style='background: #fff3e0; padding: 6px 12px; border-radius: 20px; font-size: 12px;'>🗺 خريطة تفاعلية</span>
    </div>
    <a href='viewer_full.html' target='_blank' 
       style='background: #28a745; color: white; padding: 14px 28px; text-decoration: none; border-radius: 8px; font-weight: bold; display: inline-block; box-shadow: 0 4px 12px rgba(40,167,69,0.3);'>
       🏗 فتح الفيوور الكامل
    </a>
</div>
"""))
````
